```python
import numpy as np
import csv
numpy_array = np.genfromtxt('cleaned_marketing_data (2).csv', dtype=np.int64, delimiter=",", skip_header=1)
numpy_array = np.delete(numpy_array, 0, axis=1)
numpy_array
```

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(211)
plt.title('education vs income')
plt.scatter(numpy_array[:,1], numpy_array[:,3],marker='o', alpha=0.1)
plt.show()

plt.subplot(212)
plt.title('country vs numberofweb purches')
#plt.xlable('number of children')
plt.xlabel('country')
plt.ylabel('number of web purches')
plt.scatter(numpy_array[:,11], numpy_array[:,9],marker='o', alpha=0.1)
plt.show()

plt.subplot(221)
plt.title('marital status vs amount in wine')
#plt.xlable('number of children')
plt.ylabel('amount in wine')
plt.xlabel('marital status')
plt.scatter(numpy_array[:,2], numpy_array[:,4],marker='o', alpha=0.1)
plt.show()

plt.subplot(222)
plt.title('marital status vs total spend amount')

plt.ylabel('amount soend ')
plt.xlabel('marital status')
plt.scatter(numpy_array[:,2], np.sum(numpy_array[:,4:9],axis=1),marker='*', alpha=0.3)
plt.show()
```
